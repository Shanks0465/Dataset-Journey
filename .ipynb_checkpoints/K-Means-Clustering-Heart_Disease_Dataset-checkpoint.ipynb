```python
#Import pandas,numpy for Dataset Manupilation and matplotlib and seaborn for Visualization
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```

```python
#Import functions for Model, Dataset Splitting and Evaluation
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
df=pd.read_csv('Datasets/heart.csv') #Read the Dataset CSV File to a dataframe object
```

```python
df.describe() # Further Statistical Information about the dataset
```

```python
# Display BloodPressure against BMI
df.plot(kind="scatter", x="age", y="trestbps") # Plot the data points (x-Sepal Length and y-Sepal Width)
plt.show()
print("From df.describe and the plot we can see that few rows have 0 as value for some columns")
```

```python
# Let's replace those 0 values with the mean of column
zerocols=['exang','oldpeak','slope','ca','thal','fbs','restecg','cp']
for col in zerocols:
    df[col]=df[col].replace(0,df[col].mean())
```

```python
df.describe()
```

```python
sns.FacetGrid(df, hue="target", height=5).map(plt.scatter, "age", "trestbps").add_legend() 
plt.show()
```

```python
corr=df.corr() #Correlation Matrix
```

```python
# Display the correlation matrix using a heatmap
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
# Exactly 1. A perfect downhill (negative) linear relationship

# 0.70. A strong downhill (negative) linear relationship

# 0.50. A moderate downhill (negative) relationship

# 0.25. A weak downhill (negative) linear relationship

# 0. No linear relationship


# +0.25. A weak uphill (positive) linear relationship

# +0.50. A moderate uphill (positive) relationship

# +0.70. A strong uphill (positive) linear relationship

# Exactly +1. A perfect uphill (positive) linear relationship
```

```python
X=df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']].values
y=df['target']
```

```python
# Create the training and test sets using 0.2 as test size (i.e 80% of data for training rest 20% for model testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
kmeans= KMeans(n_clusters=3,max_iter=600)
```

```python
kmeans=kmeans.fit(X_train,y_train)
```

```python
y_pred = kmeans.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Get the confusion Matrix of the Model
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
# Plot the Confusion Matrix as a HeatMap
class_names=[0,1.2] # Name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print(metrics.classification_report(y, kmeans.predict(X),zero_division=1)) # Displays a comprehensive Report of the Decision Tree Model
```
